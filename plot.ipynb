```python
scores = np.random.normal(70, 10, 1000)
```

```python
plt.hist(scores, bins=20, color='skyblue')
plt.title("Exam Scores (Gaussian Distribution)")
plt.xlabel("Scores")
plt.ylabel("Number of Students")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

visits = np.random.poisson(5, 365)

# Plot histogram
plt.hist(visits, bins=10, color='orange')
plt.title("Daily Website Visits (Poisson Distribution)")
plt.xlabel("Visits per Day")
plt.ylabel("Number of Days")
plt.show()
```

```python

names = ['Alice', 'Bob', 'Charlie', 'David', 'Eve']  # PYTHONIC APPORACH
ages = [25, 30, 35, 40, 45]

plt.bar(names, ages, color='Red')
plt.xlabel("Names")
plt.ylabel("Ages")
plt.show()
```

```python
import numpy as np                        # PYTHONIC APPORACH USING NUMPY ARRAYS
names_array = np.array(names)
ages_array = np.array(ages)

plt.bar(names_array, ages_array, color='lightblue')
plt.show()
```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

names = ['Alice', 'Bob', 'Charlie']
ages = [25, 30, 22]

df = pd.DataFrame({'Name': names, 'Age': ages})

ax = sns.barplot(x='Name', y='Age', hue='Name', data=df, palette='viridis', legend=False)
plt.title("Ages of People")
plt.show()
```
