```python
import numpy as np
import matplotlib.pyplot as plt

visits = np.random.poisson(5, 365)

# Plot histogram
plt.hist(visits, bins=10, color='orange')
plt.title("Daily Website Visits (Poisson Distribution)")
plt.xlabel("Visits per Day")
plt.ylabel("Number of Days")
plt.show()
```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = {
    "Year": [2004,2005,2006,2007,2008,2009],
    "Sales": [100,150,200,250,300,200]
}
df = pd.DataFrame(data)
df
```

```python
# using matplot

plt.figure(figsize=(10,5))
plt.bar(df['Year'], df['Sales'], color='red')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Sales Over Year')
plt.show()
```

```python
# using seaborn
plt.figure(figsize=(10,5))
sns.barplot(x='Year',y='Sales',data=df)
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Sales Over Year')
plt.show()
```

```python
# using matplot

plt.plot(df['Year'], df['Sales'], color='red', linestyle='-',marker='*')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Sales Over Year')
plt.show()
```

```python

plt.figure(figsize=(10,5))
sns.lineplot(x='Year',y='Sales',data=df, linestyle='--',marker='o')
plt.grid(True)
```

```python
# histogram using Matplot
import numpy as np
data = np.random.normal(loc=50, scale=10, size=200)

df = pd.DataFrame({'Score':data})
df
```

```python
plt.figure(figsize=(10,5))
plt.hist(df['Score'], bins=20,color='skyblue', edgecolor='black')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.title('Sales Over Year')
plt.grid(True)
plt.show()
```

```python

sns.histplot(data=df,x='Score',bins=20,kde=True, color='Salmon')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.title('Sales Over Year')
plt.grid(True)
plt.show()
```

```python
# Boxplot
math_scores = [64,70,75,80,85,90,95,100,200]
english_scores = [55,60,65,70,75,80,85,90,95]

plt.figure(figsize=(10,5))
plt.boxplot([math_scores,english_scores],labels=['Math', 'English'])
plt.xlabel('Subject')
plt.ylabel('Scores')
plt.title('Boxplot of Math and English Scores')
plt.grid(True)
plt.show()
```

```python
# using seaborn
df = pd.DataFrame({
    'subject': ["Math"]*7 + ["Science"]*7,
    'score': [88,92,80,89,100,76,95,85,92,30,40,60,70,80]
})
sns.boxplot(x="subject", y="score", data=df)
plt.title('Seaborn Box Plot')
plt.show()
```

```python
# Heatmap
data = np.array([[30,20,40],
                [25,35,30],
                [20,25,45]]
)
df= pd.DataFrame(data, columns=["Math","Science","English"], 
                 index=["class_a", "class_b", "class_c"])

# using Matplotlib
plt.imshow(data, cmap="YlGnBu")
plt.title("Heatmap")
plt.xticks([0,1,2], df.columns)
plt.yticks([0,1,2], df.index)
plt.colorbar(label='scores')
plt.show()
```
